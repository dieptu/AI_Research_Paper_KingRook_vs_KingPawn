```python
import requests
import zipfile
import io
import pandas as pd

# get the dataset
URL = 'https://archive.ics.uci.edu/static/public/22/chess+king+rook+vs+king+pawn.zip'
# Send an HTTP GET request to the URL
response = requests.get(URL)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Create a file-like object from the response content
    zip_file = io.BytesIO(response.content)

    # Extract the zip file
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall('extracted_data')

    # Assuming the extracted file is a CSV, adjust the file name accordingly
    dataset_path = 'extracted_data/kr-vs-kp.data'

    # Load the dataset into a Pandas DataFrame
    df = pd.read_csv(dataset_path, header=None)  # Adjust header parameter as needed

    # Display the first few rows of the dataset
    print(df.head())
else:
    print("Failed to download the dataset. Status code:", response.status_code)
```

```python
newColumnName = ['bkblk','bknwy','bkon8','bkona','bkspr','bkxbq','bkxcr','bkxwp','blxwp','bxqsq','cntxt','dsopp','dwipd',
 'hdchk','katri','mulch','qxmsq','r2ar8','reskd','reskr','rimmx','rkxwp','rxmsq','simpl','skach','skewr',
 'skrxp','spcop','stlmt','thrsk','wkcti','wkna8','wknck','wkovl','wkpos','wtoeg']

if len(newColumnName) == len(df.columns[:36]):
  df.columns = newColumnName + list(df.columns[36:])
else:
  print("# columns do not match")
```
